```python
import numpy as np
import matplotlib.pyplot as plt
""" Problem 2: Create 2-dimensional position vectors and plot them. """
origin = np.array([0,0])
V = np.array([1,1])
W = np.array([-2.0,-1])
fig, ax = plt.subplots()
# Add vectors V and W to the plot
ax.quiver(origin[0], origin[1], V[0], V[1], angles='xy', scale_units='xy', scale=1, color='r')
ax.quiver(origin[0], origin[1], W[0], W[1], angles='xy', scale_units='xy', scale=1, color='b')
ax.text(origin[0]-0.1, origin[1]+0.1, "O")
ax.text(V[0], V[1], "v")
ax.text(W[0]-0.125, W[1]-0.1, "w")
ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_xlabel('X')
ax.set_ylabel('Y')
plt.grid()
plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np

U = np.array([1,0.25,0])
V = np.array([0,1,0.25])
W = np.array([0.25,0,1])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

origin = [0,0,0]

ax.quiver(*origin, *U)
ax.quiver(*origin, *V)
ax.quiver(*origin, *W)

plt.show()

```

```python
def magnitude(v):
    return np.sqrt(np.sum(v**2))

U = np.array([1.5,1.5,2.0])
print(magnitude(U))
```

```python
def direction_cosines(v):
    mag = magnitude(v)
    return v / mag

O = np.array([1.5,1.5,2.0])
print(direction_cosines(O))
```
