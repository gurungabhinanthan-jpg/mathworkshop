Task 1:

Plot the graph for linear system b and write the point of intersection between three lines.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate x values
x = np.linspace(-5, 5, 100)

# Equations:
# 2x + 3y = 3  → y = (3 - 2x)/3
# 4x + 6y = 6  → y = (6 - 4x)/6 (same line as above)
# 3x - 2y = 1  → y = (3x - 1)/2

y1 = (3 - 2*x) / 3
y2 = (6 - 4*x) / 6
y3 = (3*x - 1) / 2

# Create plot
fig, ax = plt.subplots()

# Plot lines
ax.plot(x, y1, color='r')
ax.plot(x, y2, color='g', linestyle='--')  # same line (dashed)
ax.plot(x, y3, color='b')

# Label equations
ax.text(-4, 3, '$2x + 3y = 3$', color='r')
ax.text(-4, 2, '$4x + 6y = 6$', color='g')
ax.text(1, 4, '$3x - 2y = 1$', color='b')

# Intersection point
x_int = 9/13
y_int = 7/13
ax.plot(x_int, y_int, 'ko')  # black dot
ax.text(x_int, y_int, f'({x_int:.2f}, {y_int:.2f})')

# Axes settings
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.axvline(color='k', linewidth=1)
ax.axhline(color='k', linewidth=1)

ax.set_xticks(range(-5, 6))
ax.set_yticks(range(-5, 6))

ax.set_aspect('equal')
ax.grid(True, linestyle=':')

plt.xlabel("X")
plt.ylabel("Y")
plt.title("Graph of Linear System (b)")
plt.show()
```

3.	Create a 4x4 matrix containing random numbers in the range of 1 to 20. Have the user prompted for a number, check that the imputed value is a number and then check if the number is in the matrix. If the number is in the matrix, return the number of times that number is in the matrix.

```python
import numpy as np

# Create a 4x4 matrix with random numbers from 1 to 20
matrix = np.random.randint(1, 21, (4, 4))

print("Generated 4x4 Matrix:")
print(matrix)

# Take user input
user_input = input("\nEnter a number to search: ")

# Check if input is a valid number
if user_input.isdigit():
    num = int(user_input)

    # Count occurrences
    count = np.count_nonzero(matrix == num)

    if count > 0:
        print(f"\nNumber {num} found {count} time(s) in the matrix.")
    else:
        print(f"\nNumber {num} is not present in the matrix.")
else:
    print("\nInvalid input! Please enter a valid number.")
```

1.	Create a matrix 𝑨 of order 5 × 5 with random entries. Calculate the rank of the matrix 𝑨 to check the number of linearly independent columns or rows and to check if the matrix is invertible. What are the other ways to check if the matrix is invertible?


```python
import numpy as np

# Create a 5x5 matrix with random values (1–20)
A = np.random.randint(1, 21, (5, 5))

print("Matrix A:")
print(A)

# Calculate rank
rank = np.linalg.matrix_rank(A)
print("\nRank of matrix A:", rank)

# Check if invertible
if rank == 5:
    print("Matrix is invertible.")
else:
    print("Matrix is NOT invertible.")
```

2.	Generate a matrix 𝑨 of order 5 × 5 with random entries and generate a vector (column) 𝒙 of length 5 with random entries. Calculate Matrix-vector product 𝒃 = 𝑨𝒙. Now, we have a linear system, i.e. the matrix 𝑨 and a vector 𝒃. Solve this Linear System.

```python
import numpy as np

# Generate random 5x5 matrix A
A = np.random.randint(1, 21, (5, 5))

# Generate random vector x (5x1)
x = np.random.randint(1, 21, (5, 1))

# Compute b = Ax
b = np.dot(A, x)

print("Matrix A:")
print(A)

print("\nVector x:")
print(x)

print("\nVector b = Ax:")
print(b)

# Solve the system Ax = b
# (Find x back from A and b)
try:
    x_solution = np.linalg.solve(A, b)
    print("\nSolution of the system (x):")
    print(x_solution)
except np.linalg.LinAlgError:
    print("\nMatrix A is not invertible, system cannot be solved uniquely.")
```

3.	Solve the following system.
8x+3y−2z=9
−4x+7y+5z==15
3x+4y−12z=35


```python
import numpy as np

# Coefficient matrix A
A = np.array([
    [8, 3, -2],
    [-4, 7, 5],
    [3, 4, -12]
])

# Constant vector b
b = np.array([9, 15, 35])

# Solve Ax = b
solution = np.linalg.solve(A, b)

# Display result
print("Solution (x, y, z):")
print(solution)
```

Task 4:  
1.	Consider the following 10 points in X-Y plane: (−3, −7.6), (−2,1), (−1, 4.8), (0, 5), (1, 2.8), (2, −0.6), (3, −4), (4, −6.2), (5, −6), (7, 6.4). Plot these points in X-Y plane. Connect the points to guess the polynomial. Once you guess your polynomial, formulate a Linear System and then solve it to find the coefficient of the polynomial. Finally, check if the polynomial fits along with the above given points.


```python
import numpy as np
import matplotlib.pyplot as plt

# Given points
x = np.array([-3, -2, -1, 0, 1, 2, 3, 4, 5, 7])
y = np.array([-7.6, 1, 4.8, 5, 2.8, -0.6, -4, -6.2, -6, 6.4])

# Plot points
plt.scatter(x, y, color='red', label='Data Points')

# Try fitting a polynomial (degree 3 or 4 usually works well)
degree = 3
coeffs = np.polyfit(x, y, degree)

# Create polynomial function
poly = np.poly1d(coeffs)

# Generate smooth curve
x_fit = np.linspace(min(x), max(x), 200)
y_fit = poly(x_fit)

# Plot polynomial curve
plt.plot(x_fit, y_fit, label=f'Polynomial (degree {degree})')

# Show graph
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Polynomial Fit for Given Points")
plt.legend()
plt.grid(True)
plt.show()

# Print polynomial
print("Polynomial coefficients:")
print(coeffs)

print("\nPolynomial equation:")
print(poly)

# Check fit (compare actual vs predicted)
y_pred = poly(x)
print("\nActual vs Predicted:")
for i in range(len(x)):
    print(f"x={x[i]}, Actual={y[i]}, Predicted={y_pred[i]:.2f}")
```
